````python
import matplotlib.pyplot as plt
import seaborn as sns

# Calcular a volatilidade média (Amplitude média)
volatilidades = {nome: df["Amplitude"].mean() for nome, df in dados.items()}

# Converter para dataframe
df_vol = pd.DataFrame.from_dict(volatilidades, orient='index', columns=['Amplitude Média'])

plt.figure(figsize=(10,6))
sns.barplot(x=df_vol.index, y=df_vol["Amplitude Média"], palette="viridis", hue=df_vol.index, legend=False)
plt.title("Volatilidade Média (Amplitude) por Ativo")
plt.xlabel("Ativo")
plt.ylabel("Amplitude Média (High - Low)")
plt.show()
````

````python
mais_volatil = max(volatilidades, key=volatilidades.get)
menos_volatil = min(volatilidades, key=volatilidades.get)

print(" ANÁLISE DE VOLATILIDADE\n")
print(f"Ativo mais volátil: {mais_volatil} ({volatilidades[mais_volatil]:.4f})")
print(f"Ativo menos volátil: {menos_volatil} ({volatilidades[menos_volatil]:.4f})")
````

````python
mais_volatil = df_vol['Amplitude Média'].idxmax()
menos_volatil = df_vol['Amplitude Média'].idxmin()

plt.figure(figsize=(10,6))

cores = ["darkred" if ativo == mais_volatil
         else "darkgreen" if ativo == menos_volatil
         else "lightgray"
         for ativo in df_vol.index]

sns.barplot(x=df_vol.index, y=df_vol["Amplitude Média"], palette=cores, hue=df_vol.index, legend=False)

plt.title("Ativo Mais e Menos Volátil")
plt.xlabel("Ativo")
plt.ylabel("Amplitude Média (High - Low)")
plt.text(df_vol.index.tolist().index(mais_volatil),
         df_vol.loc[mais_volatil]["Amplitude Média"],
         "  MAIS VOLÁTIL", color="darkred", weight="bold")

plt.text(df_vol.index.tolist().index(menos_volatil),
         df_vol.loc[menos_volatil]["Amplitude Média"],
         "  MENOS VOLÁTIL", color="darkgreen", weight="bold")

plt.show()

print(" Resultado:")
print(f"Ativo mais volátil: {mais_volatil}")
print(f"Ativo menos volátil: {menos_volatil}")
````

Participação da volatilidade


````python
import matplotlib.pyplot as plt

# cálculo da média da volatilidade
volatilidades = {nome: df["Amplitude"].mean() for nome, df in dados.items()}

# Definir um novo conjunto de cores para o gráfico de pizza
cores_novas = ['#66c2a5', '#fc8d62', '#8da0cb'] # Verde-azulado, laranja, azul-púrpura

# Calcular porcentagens para identificar fatias pequenas e criar um 'explode'
total_vol = sum(volatilidades.values())
percentages = {k: (v / total_vol) * 100 for k, v in volatilidades.items()}

explode = [0.0] * len(volatilidades)
for i, (ativo, perc) in enumerate(percentages.items()):
    if perc < 1: # Se a porcentagem for menor que 1%, explode um pouco a fatia
        explode[i] = 0.2

plt.figure(figsize=(12,10)) # Aumentar o tamanho do gráfico para melhor legibilidade e espaço para a legenda

wedges, texts, autotexts = plt.pie(volatilidades.values(),
                                   autopct='%1.1f%%',
                                   startangle=140,
                                   colors=cores_novas,
                                   explode=explode, # Aplicar o explode
                                   pctdistance=0.85, # Ajustar a distância das porcentagens do centro
                                   textprops={'color': 'black', 'fontsize': 10}) # Definir a cor do texto das porcentagens para preto e tamanho menor

# Adicionar a legenda ao lado do gráfico de pizza
plt.legend(wedges, volatilidades.keys(),
           title="Ativos",
           loc="center left",
           bbox_to_anchor=(1, 0.5), # Posicionar a legenda à direita do gráfico
           fontsize=12,
           title_fontsize=14)

plt.title("Participação da Volatilidade Média por Ativo", fontsize=16) # Aumentar o tamanho da fonte do título
plt.axis('equal') # Garante que o gráfico de pizza seja um círculo
plt.tight_layout() # Ajustar layout para evitar que a legenda seja cortada
plt.show()
````

````python
from ta.trend import EMAIndicator, MACD
from ta.momentum import RSIIndicator
from ta.volatility import BollingerBands
````

````python
#CRIAR TODOS OS INDICADORES PARA CADA ATIVO
for nome, df in dados.items():

    # Médias móveis
    df["MM20"]  = df["Close"].rolling(20).mean()
    df["MM50"]  = df["Close"].rolling(50).mean()
    df["MM200"] = df["Close"].rolling(200).mean()

    # RSI 14 dias
    df["RSI14"] = RSIIndicator(df["Close"], window=14).rsi()

    # MACD
    macd = MACD(close=df["Close"], window_slow=26, window_fast=12, window_sign=9)
    df["MACD"]       = macd.macd()
    df["MACD_Sinal"] = macd.macd_signal()
    df["MACD_Hist"]  = macd.macd_diff()

    # Bandas de Bollinger
    boll = BollingerBands(close=df["Close"], window=20, window_dev=2)
    df["BB_Media"]    = boll.bollinger_mavg()
    df["BB_Superior"] = boll.bollinger_hband()
    df["BB_Inferior"] = boll.bollinger_lband()

print("Indicadores técnicos criados com sucesso!")
````
